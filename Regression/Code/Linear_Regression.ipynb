## 线性回归
使用了最小二乘法

```python
class linear_regression:
    def _init_(self):
        self.theta = []
        self.b = None
        self.a = []
    
    def get_parameter(self, X_train, Y_train):
        if X_train.shape[0] != Y_train.shape[0]:
            print("数据发生错误！")
            return
        ones = np.ones((X_train.shape[0] ,1))
        X_b = np.hstack([ones, X_train])
        self.theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y_train)
        self.b = self.theta[0]
        self.a = self.theta[1:]
        return self
        
    def predict(self, X_test):
        if len(self.theta) == 0:
            print("数据发生错误！")
            return
        if X_test.shape[1] != len(self.a):
            print("数据发生错误！")
            return
        ones = np.ones((X_test.shape[0] ,1))
        X_b = np.hstack([ones, X_test])
        Y_train = X_b.dot(self.theta)
        return Y_train
```

```python
reg = linear_regression()
reg
```

```python
reg.get_parameter(data, target)
results = reg.predict(data)
results
```

### 尝试求了一下平均方差

```python
sum = 0
for i in range(len(results)):
    sum = sum + (target[i] - results[i])** 2
MSE = (1/len(results))*sum
MSE
```
