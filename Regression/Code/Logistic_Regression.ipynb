```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

## 导入数据集

```python
cancer = load_breast_cancer()
cancer
```

```python
data = cancer.data
target = cancer.target
```

## 逻辑回归

```python
class Logical_regression:
    def _init_(self):
        self.alpha = None
        self.loop = None
        self.e = None
        self.X_b = None
        self.Y_train = None
        self.theta = None
        
    def sigmod(self, z):
        return 1 / (1 + np.exp(-z))
    
    #参数设置
    def get_parameter(self, alpha, loop, X_train, Y_train):
        #学习率alpha
        self.alpha = alpha
        #循环次数
        self.loop = loop
        self.Y_train = np.mat(Y_train).transpose()
        ones = np.ones((X_train.shape[0] ,1))
        X_b = np.hstack([ones, X_train])
        self.X_b = np.mat(X_b)
        self.theta = np.ones((X_b.shape[1], 1))
        return self
    
    #梯度下降法求参
    def grad_descent(self):
        if len(self.theta) != 0:
            for i in range(self.loop):
                #x = (self.X_b.dot(self.theta)).flatten()
                h = self.sigmod(self.X_b.dot(self.theta))
                self.theta = self.theta - self.alpha * self.X_b.transpose().dot(h - self.Y_train)
                theta = self.theta
            return theta
        else:
            print('发生错误！')
    
    #预测
    def predict(self, X_test):
        results = []
        ones = np.ones((X_test.shape[0] ,1))
        X_b = np.hstack([ones, X_test])
        h = self.sigmod(X_b.dot(self.theta))
        for i in range(len(h)):
            if h[i] >= 0.5:
                results.append(1)
            else:
                results.append(0)
        return results
```

```python
reg = Logical_regression()
reg.get_parameter(0.01, 1000, data, target)
reg.grad_descent()
results = reg.predict(data)
```
